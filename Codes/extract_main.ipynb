```python
import xml.etree.ElementTree as ET 
import re 
import spacy
```

## EXTRACTION TRAIN 

```python
data_train_text_it = []
data_train_target_it = []

data_train_text_en = []
data_train_target_en = []

data_train_text_fr = []
data_train_target_fr = []


################################ ANGLAIS ########################################################################

tree= ET.parse('Corpusdapprentissage/deft09_parlement_appr_en.xml')

root= tree.getroot()
doc= root.findall('doc')

for doc in root.findall('doc') :
    parti= doc.find(".//PARTI")
    val= parti.get("valeur")
    data_train_target_en.append(val)

    texte = doc.find('texte')
    if texte is not None:
        textesanspar = re.sub(r'\xa0|\n|<texte>|</texte>|<p>|</p>', '', ET.tostring(texte, 'utf-8').decode('utf-8'))
        textesanspar = re.sub(r'[^\w\s\']|[0-9]', ' ', textesanspar)

        data_train_text_en.append(textesanspar)       
        
############################### FRANCAIS #######################################################################

tree= ET.parse('Corpusdapprentissage/deft09_parlement_appr_fr.xml')

root= tree.getroot()
doc= root.findall('doc')

for doc in root.findall('doc') :
    parti= doc.find(".//PARTI")
    val= parti.get("valeur")
    data_train_target_fr.append(val)

    texte = doc.find('texte')
    if texte is not None:
        textesanspar = re.sub(r'\xa0|\n|<texte>|</texte>|<p>|</p>', '', ET.tostring(texte, 'utf-8').decode('utf-8'))
        textesanspar = re.sub(r'[^\w\s\']|[0-9]', ' ', textesanspar)

        data_train_text_fr.append(textesanspar)
        
################################## ITALIEN ###############################################################

tree= ET.parse('Corpusdapprentissage/deft09_parlement_appr_it.xml')

root= tree.getroot()
doc= root.findall('doc')

for doc in root.findall('doc') :
    parti= doc.find(".//PARTI")
    val= parti.get("valeur")
    data_train_target_it.append(val)

    texte = doc.find('texte')
    if texte is not None:
        textesanspar = re.sub(r'\xa0|\n|<texte>|</texte>|<p>|</p>', '', ET.tostring(texte, 'utf-8').decode('utf-8'))
        textesanspar = re.sub(r'[^\w\s\']|[0-9]', ' ', textesanspar)

        data_train_text_it.append(textesanspar)
        
```
